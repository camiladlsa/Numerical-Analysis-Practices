### Actividad 4

```python
# Número de monedas:

def calcular_distribucion_optima(importe):
    monedas_disponibles = [25, 10, 5, 1]
    distribucion = [0, 0, 0, 0]  # Inicializamos el vector de distribución a ceros

    # Iteramos sobre las monedas disponibles
    for i in range(len(monedas_disponibles)):
        # Mientras el importe sea mayor o igual al valor de la moneda actual
        while importe >= monedas_disponibles[i]:
            distribucion[i] += 1  # Incrementamos la cantidad de esa moneda
            importe -= monedas_disponibles[i]  # Restamos el valor de la moneda al importe

    return distribucion

# Solicitar el importe al usuario

while True:

    importe = int(input("Por favor ingresar un importe entre 1 y 100 pesos: "))

    if 1 <= importe <= 100:
        distribucion = calcular_distribucion_optima(importe)
        print("\nDistribución óptima de monedas:\n")
        print(f"Monedas de 25 pesos: {distribucion[0]}")
        print(f"Monedas de 10 pesos: {distribucion[1]}")
        print(f"Monedas de 5 pesos: {distribucion[2]}")
        print(f"Monedas de 1 peso: {distribucion[3]}")
        break
    else:
        print("\nEl importe debe estar entre 1 y 100 pesos... ¡Intente nuevamente!\n")
```

### Actividad 5

```python
# Algoritmo de Herón de Alejandría: 

def heron_aprox_raiz(num, g, tolerancia):
    x = g  # Estimación inicial
    while True:
        siguiente_x = 0.5*(x+num/x)  # Fórmula de Herón
        if abs(siguiente_x-x)<tolerancia:
            break  # Si la diferencia entre las estimaciones es menor que la tolerancia, se termina el programa
        x = siguiente_x
    return x

num = int(input("Ingrese el número cuya raíz desea aproximar: "))
g = 6 # Estimación inicial planteada
tolerancia = 0.0001

raiz_aprox = heron_aprox_raiz(num, g, tolerancia)

print(f"\nLa aproximación de √{num} con el Algoritmo de Herón de Alejandría es: {raiz_aprox}")
```

```python
# Bisección:

import math as m

def biseccion(ecuacion, a, b, n):

    for i in range(n):
        c = (a + b) / 2  # Punto medio del intervalo

        if ecuacion(c) == 0 or abs(b - a) < 1e-10: # Verificar la presición del resultado según la tolerancia dada.
            return c
        
        # Actualización del intervalo
            
        if ecuacion(a) * ecuacion(c) < 0:
            b = c  # Nuevo intervalo [a, c]
        else:
            a = c  # Nuevo intervalo [c, b]

    return (a + b) / 2

def ecuacion(x):
    return m.exp(x)-5*x 

a = 0
b = 2
n = 10

raiz = biseccion(ecuacion, a, b, n)

print(f"El valor de la raíz es: {raiz}")
```

### Actividad 14

```python
# Números binarios a decimal:

def bin2dec(binario):
    
    decimal = 0
    
    for i, bit in enumerate(binario[::-1]):
        if bit == '1':
            decimal += 2**i
            
    return decimal

num_binario = input("Ingrese un número binario que desea convertir a decimal: ")

num_decimal = bin2dec(num_binario)

print(f"\nEl número binario {num_binario} es equivalente al número decimal {num_decimal}")
```

### Actividad 15

```python
# Números decimales a binario:
    
def dec2bin(decimal):
    
    binario = ""
    
    while decimal > 0:
        binario = str(decimal % 2) + binario
        decimal = decimal // 2
        
    return binario

num_binario = int(input("Ingrese el número binario que desea representar a decimal: "))

num_decimal = dec2bin(num_binario)

print(f"\nEl número decimal {num_decimal} es equivalente al número binario {num_binario}")
```

```python
# Probabilidad de fallo mecánico:
    
import numpy as np
import matplotlib.pyplot as plt

def simular_fallos():
    fallo = [0, 1, 2]
    probabilidad = [0.85, 0.1, 0.05]
    cantidad = int(np.random.choice(fallo, 1, p = probabilidad))
    return cantidad

rango = np.arange(1, 101, 1)

fallos_diarios = []
fallos_acumulados = []
acumulado = 0

for i in range(100):
    fallos = simular_fallos()
    fallos_diarios.append(fallos)
    acumulado += fallos
    fallos_acumulados.append(acumulado)

tabla = np.array([rango, fallos_diarios])

np.savetxt("ejecuciones.txt", np.transpose(tabla))
load = np.loadtxt("ejecuciones.txt")

#Gráfica de los fallos diarios

plt.plot(rango, fallos_diarios)
plt.xlabel('Día')
plt.ylabel('Fallos diarios')
plt.title('Fallos diarios')
plt.show()

# Gráfica de los fallos acumulados

plt.plot(rango, fallos_acumulados)
plt.xlabel('Día')
plt.ylabel('Fallos acumulados')
plt.title('Fallos acumulados')
plt.show()
```

```python
# Criba de Eratóstenes:

def criba_eratostenes(n_max):

    numeros = list(range(2, n_max+1))
    primos = []

    while numeros:
        primo = numeros[0] # Tomar el primer número en la lista (siempre será primo)
        primos.append(primo)

        numeros = [x for x in numeros if x % primo != 0] # Tachar todos los múltiplos de ese número

    return primos

val = criba_eratostenes(1000)

print(f"Los números primos existentes hasta 1000 son: \n\n{val}")
```
